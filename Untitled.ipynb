```python
x = [1,2,3,4,5,6,7,8]
y = np.exp(x)

# Create empty figure
fig = plt.figure()

# Create axes

ax = plt.axes()

# Plot on that axes

ax.plot(x, y, 'r') # green color

# Add the title 

ax.set_title('This is a title')

# Add X and Y label

ax.set_xlabel('This is X Label')
ax.set_ylabel('This is Y Label')
```

```python
# Create empty figure
#fig = plt.figure()

# Create axes
fig, axes = plt.subplots(nrows=2, ncols=1)

# Plot on that axes

axes[0].plot(x, y, 'g') # green color
axes[1].plot(x, y**2, 'o', 'r') # blue color

# Add the title 

axes[0].set_title('This is a title 1')
axes[1].set_title('This is a title 2')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('This is Xlabel')
    ax.set_ylabel('This is Y Label')
    ax.set_title('This is a Title')

# Show the figure

fig

# Use plt.tight_layout to adjust the positions of the axes, and labels automatically 
# Comment it and see how bad the figure is
#plt.tight_layout()
```
